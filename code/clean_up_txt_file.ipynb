```python
import requests,re,os
os.chdir("..")
data_dir="data/fiscal_txt"
out_dir="data/text_clean"

doc_list=os.listdir(data_dir)
for doc in doc_list:
    with open(os.path.join(data_dir,doc),"r") as mf:
        data=mf.readlines()
    indices=[-1]+[i for i,x in enumerate(data) if x=="\n"]+[len(data)]
    output=[]
    for i in range(len(indices)-1):
        temp=[k[:-1] for k in data[indices[i]+1:indices[i+1]]]
        temp1=" ".join(temp)
        temp=temp1.replace("U.S.","United States").replace("U.K.","United Kingdom").replace("e.g.","for example").replace(";",".").replace("medium term","medium-term").replace("mediumterm","medium-term")
        temp = re.sub("\.\d+ ",".",temp)
        temp = re.sub("\(i+\)","",temp)
        output.append(temp)
        
    outfile=open(os.path.join(out_dir,doc),"w")
    for item in output:
        outfile.write("%s " %item)
```
