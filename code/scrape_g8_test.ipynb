```python
for key, value in link.items():
    html = requests.get(value)
    soup = BeautifulSoup(html.content,"lxml")
    out = soup.findAll("li",text=re.compile("fiscal"))
    out1 = soup.findAll("p",text=re.compile("fiscal"))
    temp = []
    save_dir = os.path.join(datadir,key+".txt")
    if len(out)>0:
        for k in out:
            temp1=re.sub(re.compile(r"\d+. "),"",k.contents[0])
            temp.append(temp1)
    elif len(out1)>0:
        for k in out1:
            temp1=re.sub(re.compile(r"\d+. "),"",k.contents[0])
            temp.append(temp1)
    with open(save_dir, 'w') as f:
        f.write(" ".join(temp))
```
